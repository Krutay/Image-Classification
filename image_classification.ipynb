```python
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):
  def __init__(self):
    super(CNN,self).__init__()
    self.conv1 = nn.Conv2d(3,32,kernel_size = 3)
    self.conv2 = nn.Conv2d(32,64,kernel_size = 3)
    self.conv3 = nn.Conv2d(64,96,kernel_size=3)
    self.conv4 = nn.Conv2d(96,96, kernel_size = 3)

    self.conv5 = nn.Conv2d(96,64, kernel_size = 3)

    self.dropout_conv = nn.Dropout2d(0.2)
    self.dropout_small = nn.Dropout(0.2)
    self.dropout_large = nn.Dropout(0.3)

    self.fc1 = None
    self.fc2 = nn.Linear(256,128)
    self.fc3 = nn.Linear(128,2)



  def forward(self,x):
    x = F.max_pool2d(F.relu(self.conv1(x)),2)
    x = F.max_pool2d(F.relu(self.conv2(x)),2)
    x = F.max_pool2d(self.dropout_conv(F.relu(self.conv3(x))),2)
    x = F.max_pool2d(self.dropout_conv(F.relu(self.conv4(x))),2)

    x = torch.flatten(x, start_dim = 1)

    if self.fc1 is None:
      self.fc1 = nn.Linear(x.shape[1],256).to(x.device)

    x = self.dropout_small(F.relu(self.fc1(x)))
    x = self.dropout_large(F.relu(self.fc2(x)))
    x = self.fc3(x)

    return x

cat_vs_dog = CNN()
print(cat_vs_dog)
```
